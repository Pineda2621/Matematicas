```python
sucesion = []
for i in range(36):
    n = i + 1
    sucesion.append(200+(n-1)*24)
    print("Término",n,":",sucesion[i])
```

```python
suma = sum(sucesion)
print("La suma de los primeros términos es",suma)
```

```python
sucesion = []

# Queremos calcular hasta el mes 9
for i in range(9):
    n = i + 1
    
    # Rellenamos el "Término general" con la fórmula:
    # Capital * (1.012) elevado a la n
    termino = 2000000 * (1.012)**n
    
    sucesion.append(termino)
    
    # Imprimimos con formato de moneda para que se lea mejor (sin decimales largos)
    print("Mes", n, ": $", "{:,.0f}".format(sucesion[i]))
```

```python
sucesion = []
for i in range(9):
    n = i + 1
    sucesion.append(533*(2**(n-1)))
    print("Término",n,":",sucesion[i])
```

```python
final = 34112
termino = 0
n = 0
dif = final-termino
while dif > 0:
    n += 1
    termino = sucesion[n-1]
    dif = final-termino

if dif==0:
    print("El término",final,"está en la posición",n)
else:
    print("El término",final,"no está en la sucesión")
```
